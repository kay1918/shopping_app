```python
def get_html(keyword):
    api = f'https://search.shopping.naver.com/search/all.nhn?query={keyword}&cat_id=&frm=NVSHATC'
    resp = requests.get(api)
    return resp.text
```

# 연관 키워드

```python
def get_recommend_keywords(tag):
    soup = BeautifulSoup(get_html(tag), 'html.parser')
    keywords = soup.find('div', {'class': 'relatedTags_relation_srh__1CleC'})
    keyword_list=[]
    for word in keywords:
        keyword_list.append(word.text)
    return keyword_list[1].split(' ')
```

```python
get_recommend_keywords('러블리')
```

# 연관 카테고리

```python
def get_recommend_categories(tag):
    soup = BeautifulSoup(get_html(tag), 'html.parser')
    if '카테고리' in soup.find('div', {'class': 'filter_finder__1Gtei'}).text:
        related = soup.find('ul', {'class': 'filter_finder_tit_list__2Cxcx'})
        category_list =[]
        for relate in related:
            category_list.append(relate.text)
        return category_list
    else:
        return []
```

# 연관 브랜드

```python
def get_recommend_brands(tag):
    soup = BeautifulSoup(get_html(tag), 'html.parser')
    if '브랜드' in soup.find('div', {'class': 'filter_finder__1Gtei'}).text:
        brands = soup.find('ul', {'class': 'filter_finder_tit_list__2Cxcx'})
        brand_list =[]
        for brand in brands:
            brand_list.append(brand.text)
        return brand_list[-1].split(' ')
    else:
        return []
```
